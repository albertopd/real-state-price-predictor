## Transforming to a toy example

```python
toy_df = raw_csv.sample()
print(f" example:  {toy_df}")
print(f"\n Actual price:  {toy_df['price']}")
```

```python
# predicted_price = 403107.94
# actual_price = 299500.0
```

```python
toy_df = toy_df[['habitableSurface', 'type', 'subtype','province',  'postCode','epcScore',
    'bedroomCount',
    'bathroomCount',
    'toiletCount',
    'terraceSurface',
    'gardenSurface',
    'hasAttic',
    'hasGarden',
    'hasAirConditioning',
    'hasArmoredDoor',
    'hasVisiophone',
    'hasTerrace',
    'hasOffice',
    'hasSwimmingPool',
    'hasFireplace',
    'hasBasement',
    'hasDressingRoom',
    'hasDiningRoom',
    'hasLift',
    'hasHeatPump',
    'hasPhotovoltaicPanels',
    'hasLivingRoom',]]
```

```python
toy_df.dtypes
```

### Transform columns

```python
#### Boolean Columns
boolean_cols = toy_df.columns[toy_df.columns.str.startswith('has')]
for col in boolean_cols:
    toy_df[col]= toy_df[col].astype("boolean")
```

```python
len(boolean_cols)
```

```python
numerical_cols = toy_df.select_dtypes(['float64','int64'])
numerical_cols.dtypes

```

```python
int_cols = toy_df.columns[toy_df.columns.str.endswith('Count')]
int_cols
```

```python
for col in int_cols:
    try:
        toy_df[col]= toy_df[col].astype("Int64")
    except TypeError as e:
        print(f"Error {e}")
```

```python
float_cols = toy_df.columns[toy_df.columns.str.endswith('Surface')]
float_cols
```

```python
for col in float_cols:
    try:
        toy_df[col]= toy_df[col].astype("Float64")
    except TypeError as e:
        print(f"Error {e}")
```

```python
string_cols = toy_df.select_dtypes(['object'])
string_cols 
```

```python
for col in string_cols :
    try:
        toy_df[col]= toy_df[col].astype("string")
    except TypeError as e:
        print(f"Error {e}")
```
